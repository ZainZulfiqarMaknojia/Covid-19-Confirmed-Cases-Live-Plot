```python
import pandas as pd
import numpy as np
```

```python
confirmedCases = pd.read_csv("Confirmed.csv")
```

```python
confirmedCases.head()
```

```python
confirmedCases = confirmedCases.drop(['Province/State','Lat','Long'],axis=1)
confirmedCases = confirmedCases.sort_values(by=['5/27/20'] , ascending = False) 
confirmedCases
```

```python
#confirmedCases = confirmedCases[confirmedCases["Country/Region"] == "Pakistan"]
confirmedCases = confirmedCases.iloc[0:26,:]
confirmedCases
```

```python
confirmedCases
```

```python
confirmedCases = confirmedCases.reset_index()
confirmedCases = confirmedCases.drop(['index'],axis=1)
confirmedCases = confirmedCases.rename(columns={'Country/Region': 'index'})
confirmedCases
```

```python
deletedColumn = confirmedCases['index']
confirmedCases = confirmedCases.drop('index',axis = 1)
confirmedCases.columns
```

```python
confirmedCases
#deletedRows = confirmedCases.iloc[:,:35]
```

```python
confirmedCases = confirmedCases.transpose()
confirmedCases
```

```python
confirmedCases.columns = deletedColumn
confirmedCases
```

```python
confirmedCases = confirmedCases.reset_index()
confirmedCases
confirmedCases = confirmedCases.rename(columns={'index': 'Date'})
confirmedCases
```

```python
countriesName = confirmedCases.iloc[0:0,1:]
CountriesName = []
for i in countriesName:
    CountriesName.append(i)
   
    
```

```python
CountriesName
```

```python
for j in range(26):
    temp = confirmedCases.iloc[:,j+1:j+2]
    newCases = []
    k = CountriesName[j]
    for i in range(len(temp)):
        if i == 0:
            newCases.append(temp[k][i])
        else:
            newCases.append(temp[k][i] - temp[k][i-1])
    newCases
    newCases = np.array(newCases)
    temp['newCases'] = newCases
    temp.to_csv(CountriesName[j]+".csv",index=False)  
```
