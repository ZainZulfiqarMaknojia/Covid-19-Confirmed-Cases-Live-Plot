# Hye Today We Will Plot Live Covid 19 Data With Pandas,Numpy And Matplotlib

```python
import pandas as pd
import numpy as np

# Here We Are Importing Pandas And Numpy With Their Alias
```

```python
confirmedCases = pd.read_csv("confirmed.csv") # We have Covid confirmed cases data in confirmed.csv which we are reading through pandas function read_csv
```

```python
confirmedCases.head() # Let examine first few lines of our data
```

```python
confirmedCases = confirmedCases.drop(['Province/State','Lat','Long'],axis=1) # here we notice that latitude and longitude are of no use so we will delete them plus region/provinces because we only need data of countries
confirmedCases = confirmedCases.sort_values(by=['5/30/20'] , ascending = False) # here we are plotting for first 15 countries so we will sort them in descending order by the latest
confirmedCases
```

```python
confirmedCases = confirmedCases.iloc[0:15,:] # here we have taken only top 15 countries with most confirmed cases
confirmedCases
```

```python
confirmedCases = confirmedCases.reset_index() # lets reset index to have clear view
confirmedCases = confirmedCases.drop(['index'],axis=1) # we will drop old index and axis = 1 means we will not use it that old index as our column
confirmedCases = confirmedCases.rename(columns={'Country/Region': 'index'}) # here we are renaming country/region to index which you will understand in a while
confirmedCases
```

```python
deletedColumn = confirmedCases['index'] 
confirmedCases = confirmedCases.drop('index',axis = 1) # we are droping index(which we renamed earlier(that was countries name so that we can make them as column name)
confirmedCases.columns #we are changing column names to that of index which is 0,1,2,3,4 ...
```

```python
confirmedCases
```

```python
confirmedCases = confirmedCases.transpose() # now we will take transpose so that data will become our index to have 0,1,2,3 .. as our column
confirmedCases # our next task is to change 0,1,2 ... with countries name that was stored in DeletedColumn above
```

```python
confirmedCases.columns = deletedColumn #here we have done it we have renamed our column name with that of countries
confirmedCases
```

```python
confirmedCases = confirmedCases.reset_index() 
confirmedCases
confirmedCases = confirmedCases.rename(columns={'index': 'Date'})
confirmedCases
#here we have reset the index change it from date to number
```

```python
df = confirmedCases['Date']
df.dtype
#df.to_csv("Date.csv") # this error is because we have run above statement twice and it will be solve by re running the program
```

```python
countriesName = confirmedCases.iloc[0:0,1:]
CountriesName = []
for i in countriesName:
    CountriesName.append(i)
   
    
```

```python
CountriesName # here we have fetch top 15 countries name
```

```python
for j in range(15):
    temp = confirmedCases.iloc[:,j+1:j+2]
    newCases = []
    k = CountriesName[j]
    for i in range(len(temp)):
        if i == 0:
            newCases.append(temp[k][i])
        else:
            if(temp[k][i] - temp[k][i-1] < 0):
                newCases.append(0)
            else:
                newCases.append(temp[k][i] - temp[k][i-1])
    newCases
    newCases = np.array(newCases)
    temp['newCases'] = newCases
    temp.to_csv(CountriesName[j]+".csv",index=False)
    
# this whole statement is saving top 15 countries total cases and new cases in seperate csv files of their name like US.csv,china.csv    
```
